```python
from pathlib import Path
import pandas as pd

PROJECT_ROOT = Path("..")

raw_dir = PROJECT_ROOT / "data_raw"
out_dir = PROJECT_ROOT / "data_processed"
out_dir.mkdir(exist_ok=True)

files = {
    "pl_matchlist_21-22.csv": "21-22",
    "pl_matchlist_22-23.csv": "22-23",
    "pl_matchlist_23-24.csv": "23-24",
    "pl_matchlist_24-25.csv": "24-25",
}

def detect_sep(path: Path) -> str:
    first_line = path.open("r", encoding="utf-8", errors="ignore").readline()
    return ";" if first_line.count(";") > first_line.count(",") else ","

for fname, season in files.items():
    in_path = raw_dir / fname
    if not in_path.exists():
        raise FileNotFoundError(f"NÃ£o encontrei: {in_path.resolve()}")

    sep = detect_sep(in_path)
    df = pd.read_csv(in_path, sep=sep)
    df["Season"] = season

    out_path = out_dir / fname.replace(".csv", "_v1.csv")
    df.to_csv(out_path, index=False, sep=",", encoding="utf-8")

    print(f"OK -> {out_path} (lido com '{sep}', gravado com ',')")
```
